```python
# Imports
import numpy as np
import matplotlib.pyplot as plt

from math import *
from IPython import display
```

```python
# Constants

n = 100
alpha = 0.8
T = 600
sigma = 10

weights = np.random.random((100,100,3))
print(weights.shape)
# diff = np.abs(np.sum(normRGB[0] - w, axis=2))
plt.imshow(weights)
```

```python
# Data Processing

data = np.array([
    [255,0,0],
    [0,255,0],
    [0,0,255],
    [255,255,0],
    [255,0,255],
    [0,255,255],
    [128,128,0],
    [128,0,128],
    [0,128,128],
    [255,128,0],
    [255,0,128],
    [128,255,0],
    [0,255,128],
    [128,0,255],
    [0,128,255],
    [255,20,147],
    [220,20,60],
    [255,51,51],
    [255,153,51],
    [255,255,51],
    [51,255,51],
    [153,255,51],
    [51,255,153],
    [51,255,255]]
)
norm_data = data/255

plt.imshow(np.reshape(norm_data,(norm_data.shape[0],1,3)))
```

```python
for i in range(T):
    for color in norm_data:
            
        difference = np.linalg.norm(color - weights, axis =2)
        index = np.unravel_index(np.argmin(difference, axis=None), difference.shape)
        
        for weight_i in range(len(weights)):
            for weight_j in range(len(weights[weight_i])):
                
                # Calculating N,i,j(k)
                d = dist = np.linalg.norm(np.asarray(index) - np.array([weight_i, weight_j]))
                sigma_k = sigma * exp(-i / T)
                N = exp(-(d*d / (2 * sigma_k * sigma_k)))
                alpha_k = alpha * exp(-i / T)
                
                delta_weight = alpha_k*N*(color-weights[weight_i][weight_j])
                weights[weight_i][weight_j] += delta_weight
        
    plot_ind = [20, 40, 100, 600]
    if i + 1 in plot_ind:
        print("Epoch Number: {}".format(i + 1))
        plt.imshow(weights)
        display.display(plt.gcf())
    

sigma = 40
weights = np.random.random((100,100,3))

for i in range(T):
    for color in norm_data:
            
        difference = np.linalg.norm(color - weights, axis =2)
        index = np.unravel_index(np.argmin(difference, axis=None), difference.shape)
        
        for weight_i in range(len(weights)):
            for weight_j in range(len(weights[weight_i])):
                
                # Calculating N,i,j(k)
                d = dist = np.linalg.norm(np.asarray(index) - np.array([weight_i, weight_j]))
                sigma_k = sigma * exp(-i / T)
                N = exp(-(d*d / (2 * sigma_k * sigma_k)))
                alpha_k = alpha * exp(-i / T)
                
                delta_weight = alpha_k*N*(color-weights[weight_i][weight_j])
                weights[weight_i][weight_j] += delta_weight
        
    plot_ind = [20, 40, 100, 600]
    if i + 1 in plot_ind:
        print("Epoch Number: {}".format(i + 1))
        plt.imshow(weights)
        display.display(plt.gcf())
```
