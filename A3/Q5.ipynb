```python

import random
import time
import matplotlib.pyplot as plt
import numpy as np
from math import *
from scipy.integrate import quad
```

```python
S = [0, 6]
def u_a(x, lmda, n, a):
  return exp(-lmda * pow(abs(x - a), n))
```

```python
# Various a  values
lmda = 2
n = 2
a = 3

x = np.arange(0, 6, 0.1)
y_1 = np.array([u_a(i, lmda, n, 2) for i in x])
y_2 = np.array([u_a(i, lmda, n, 3) for i in x])
y_3 = np.array([u_a(i, lmda, n, 4) for i in x])
fig, ax = plt.subplots()
ax.plot(x,y_1)
ax.plot(x,y_2)
ax.plot(x,y_3)
plt.legend(["a = 2", "a = 3", "a = 4"])
```

```python
# Various l  values
n = 2
a = 3

x = np.arange(0, 6, 0.1)
y_1 = np.array([u_a(i, 1, n, a) for i in x])
y_2 = np.array([u_a(i, 2, n, a) for i in x])
y_3 = np.array([u_a(i, 10, n, a) for i in x])
fig, ax = plt.subplots()
ax.plot(x,y_1)
ax.plot(x,y_2)
ax.plot(x,y_3)
plt.legend(["l = 1", "l = 2", "l = 3"])
```
